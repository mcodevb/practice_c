```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
plt.ion()
plt.rcParams['figure.figsize'] = 10, 8
# solve the system dy/dt = f(y, t)
def f(y, t):
     f0 = y[1]
     f1 = y[1] - np.sin(t)
     return [f0, f1]

# initial conditions
x0 = [1, -1]     # initial condition vector
t  = np.linspace(-1.0, 1., 1000)         # time grid

# solve the DEs
soln = odeint(f, x0, t)
x    = soln[:, 0]

# plot results
plt.figure()
plt.plot(t, x)
plt.plot([-1, 1],[0, 0])
plt.plot(0,0)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.title('x vs. time')
plt.legend(loc=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
plt.ion()
plt.rcParams['figure.figsize'] = 10, 8
# solve the system dy/dt = f(y, t)
def f2(y, t):
     f0 = y[1]
     f1 = y[2]
     f3 = t*y[0]-y[2]
     return [f0, f1, f3]

# initial conditions
x02 = [1, -1, 2]     # initial condition vector
t  = np.linspace(-1.0, 1., 1000)         # time grid

# solve the DEs
soln = odeint(f2, x02, t)
x    = soln[:, 0]

# plot results
plt.figure()
plt.plot(t, x , label=None)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.title('x vs. time')
plt.legend(loc=0)
```
